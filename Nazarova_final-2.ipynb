#### Начало, конец, продолжительность сессий

Подготовим таблицу для удобства анализа

```python
sessions = mobile.sort_values(by=['Session', 'date_time']).groupby('Session').agg({'user':'first', 'sourse':'first', 'event': ['first', 'last', 'count'],'date_time': ['min', 'max']}).reset_index()
sessions.columns = ['Session', 'user', 'sourse', 'first_event', 'last_event', 'number_of_events', 'session_start', 'session_end']
#sessions.columns = sessions.columns.droplevel([0])
sessions['session_duration'] = sessions['session_end'] - sessions['session_start']
sessions.head()
```

Добавим столбцы с данными о событиях внутри сессий

```python
#def 
```

```python
#просмотр номера телефона - целевое действие
contacts_show = mobile.query('event == "contacts_show"')
sessions['contacts_show'] =  sessions['Session'].isin(contacts_show['Session'])
#действия связанные с поиском 
search = mobile.query('event == "search"')
sessions['search'] =  sessions['Session'].isin(search['Session'])
# Пользователи,  видевшие рекомендованные объявления
tips_show = mobile.query('event == "tips_show"')
sessions['tips_show'] =  sessions['Session'].isin(tips_show['Session'])
# Пользователи,  кликнувшие по рекламному  объявлению
tips_click = mobile.query('event == "tips_click"')
advert_open = mobile.query('event == "advert_open"')
photos_show = mobile.query('event == "photos_show"')
map_event = mobile.query('event == "map"')
favorites_add = mobile.query('event == "favorites_add"')
contacts_call = mobile.query('event == "contacts_call"')
#добавим столбец кликнувших рекламное обьявление
sessions['tips_click'] =  sessions['Session'].isin(tips_click['Session'])
#добавим столбец открыл карточки объявлени
sessions['advert_open'] =  sessions['Session'].isin(advert_open['Session'])
#добавим столбец просмотрел фотографий в объявлении
sessions['photos_show'] =  sessions['Session'].isin(photos_show['Session'])
sessions['map'] =  sessions['Session'].isin(map_event['Session'])
sessions['favorites_add'] =  sessions['Session'].isin(favorites_add['Session'])
sessions['contacts_call'] =  sessions['Session'].isin(contacts_call['Session'])
sessions.head()
```

### Проверка гипотез

#### Гипотеза 1

Одни пользователи совершают действия tips_show и tips_click, другие — только tips_show. Проверьте гипотезу: конверсия в просмотры контактов различается у этих двух групп.




```python
funnel_by_users = users.query('tips_show == True').groupby(['tips_show','tips_click' ]).agg({'user': 'nunique', 'contacts_show': 'sum'})
funnel_by_users['%'] = funnel_by_users['contacts_show']/funnel_by_users['user']
funnel_by_users
```

1. видел `tips_show` не пользовался `tips_click` -  конверсия 17%
2. видел `tips_show` пользовался `search` - конверсия 31% значительно выше

* Н0: конверсия пользователей совершающих действия tips_show и tips_click не отличатся от конверсии пользователей совершающих только tips_show
* Н1: конверсия пользователей совершающих действия tips_show и tips_click отличатся от конверсии пользователей совершающих только tips_show

```python
count = funnel_by_users['contacts_show']
nobs = funnel_by_users['user']
alpha = .05

stat, p_val = proportions_ztest(count, nobs)
print('p-значение: {0:0.3f}'.format(p_val))
if p_val < alpha:
        print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
else:
        print(
        'Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными'
    )

```

#### Гипотеза 2

Изначально предполагалась проверка гипотезы о различиях конверсии по источникам (гугл, яндекс), но в результате такая гипотеза не кажется интересной, разницы явно нет. Поэтому возьмем следующую: 

Одни пользователи совершают действия search и favorites_add, другие — только search. Проверьте гипотезу: конверсия в просмотры контактов различается у этих двух групп.


```python
funnel_2 = users.query('search == True').groupby(['search','favorites_add' ]).agg({'user': 'nunique', 'contacts_show': 'sum'})
funnel_2['%'] = funnel_2['contacts_show']/funnel_2['user']
funnel_2
```

* Н1: конверсия пользователей совершающих действия search и favorites_add отличатся от конверсии пользователей совершающих только search

```python
count = funnel_2['contacts_show']
nobs = funnel_2['user']
alpha = .05

stat, p_val = proportions_ztest(count, nobs)
print('p-значение: {0:0.3f}'.format(p_val))
if p_val < alpha:
        print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
else:
        print(
        'Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными'
    )

```
