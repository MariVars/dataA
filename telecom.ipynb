### Обьединение таблиц

```python

calls_internet = calls_month.merge(internet_month, on = ['user_id', 'month'] , how = 'outer')
calls_internet_messages = calls_internet.merge(messages_month, on = ['user_id', 'month'] , how = 'outer')
user_calls_internet_messages = df_users.merge(calls_internet_messages, on = 'user_id', how = 'left')
user_calls_internet_messages
```

### Расчет затрат по каждому тарифу

#### Тариф «Смарт»
Ежемесячная плата: 550 рублей
Включено 500 минут разговора, 50 сообщений и 15 Гб интернет-трафика
Стоимость услуг сверх тарифного пакета:
минута разговора: 3 рубля
сообщение: 3 рубля
1 Гб интернет-трафика: 200 рублей

```python
smart = user_calls_internet_messages.query('tariff == "smart"')
```

```python
def cost(row):
    if row['duration_ceil'] > 500:
        call_over = (row['duration_ceil'] - 500)*3
    else:
        call_over = 0
    if row['message_date'] > 50:
        messages_over = (row['message_date'] - 50)*3
    else:
        messages_over = 0
    if row['mb_month'] > 15:
        mb_over = (row['mb_month']-15)*200
    else:
        mb_over = 0
    cost= call_over + messages_over + mb_over +550
    return cost
smart['cost'] = smart.apply(cost, axis=1)

smart
```

####  Тариф «Ультра»
Ежемесячная плата: 1950 рублей
Включено 3000 минут разговора, 1000 сообщений и 30 Гб интернет-трафика
Стоимость услуг сверх тарифного пакета:
минута разговора: 1 рубль
сообщение: 1 рубль
1 Гб интернет-трафика: 150 рублей

```python
ultra = user_calls_internet_messages.query('tariff == "ultra"')
```

```python
def cost(row):
    if row['duration_ceil'] > 3000:
        call_over = (row['duration_ceil'] - 3000)*1
    else:
        call_over = 0
    if row['message_date'] > 1000:
        messages_over = (row['message_date'] - 1000)*1
    else:
        messages_over = 0
    if row['mb_month'] > 30:
        mb_over = (row['mb_month']-30)*150
    else:
        mb_over = 0
    cost= call_over + messages_over + mb_over +1950
    return cost
ultra['cost'] = ultra.apply(cost, axis=1)

ultra

```

Соединим таблицы по двум тарифам, удалим лишние столбцы и переименуем нужные.

```python
tariffs = pd.concat([smart, ultra]).reset_index()
tariffs = tariffs.drop(['churn_date', 'reg_date', 'mb_used'], 1)
tariffs = tariffs.rename(columns={'duration_ceil':'calls', 'mb_month':'internet', 'message_date':'sms'})
tariffs
```

## "Выгодность" тарифов

Посчитаем для каждого абонента: дороже или дешевле обошелся бы другой тариф

```python
def cost_another(tariffs):
    if  tariffs['tariff'] == 'smart':
        if tariffs['calls'] > 3000:
            call_over = (tariffs['calls'] - 3000)*1
        else:
            call_over = 0
        if tariffs['sms'] > 1000:
            messages_over = (tariffs['sms'] - 1000)*1
        else:
            messages_over = 0
        if tariffs['internet'] > 30:
            gb_over = (tariffs['internet']-30)*150
        else:
            gb_over = 0
        cost_another= call_over + messages_over + gb_over +1950
        return cost_another
    else:
        if tariffs['calls'] > 500:
            call_over = (tariffs['calls'] - 500)*3
        else:
            call_over = 0
        if tariffs['sms'] > 50:
            messages_over = (tariffs['sms'] - 50)*3
        else:
            messages_over = 0
        if tariffs['internet'] > 15:
            gb_over = (tariffs['internet']-15)*200
        else:
            gb_over = 0
        cost_another= call_over + messages_over + gb_over +550
        return cost_another

    
tariffs['cost_another'] = tariffs.apply(cost_another, axis=1)

tariffs

```

Если бы абонент пользовался другим тарифом и потратил бы меньше, значит тариф для него не выгодный. Посмотрим сколько таких 'переплативших' в каждом тарифе

```python
def benefit(tariffs):
    if tariffs['cost_another'] - tariffs['cost'] >= 0:
        benef = 'выгодно'
    else:
        benef = 'не выгодно'
    return benef
tariffs['benef'] = tariffs.apply(benefit, axis=1)
tariffs
```

```python
tariffs.pivot_table(index = 'tariff', columns = 'benef', values = 'cost', aggfunc= 'count', margins=True)
```

В тарифе "Смарт" переплачивали 19%, в тарифе "Ультра" переплачивали 50%. Возмножно Мегалайну выгоднее, чтобы больше абонентов переплачивали, но рано или поздно клиент заметит, что платит за то , что ему не нужно. Все-таки развивать нужно разные тарифы)

## Проверка гипотез

### Средняя выручка пользователей тарифов «Ультра» и «Смарт» 


Гипотезы:

Ho - средняя выручка пользователей тарифов «Ультра» и «Смарт» равны,

H₁ - средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются

```python
array_ultra = tariffs.query('tariff == "ultra"')['cost']
array_smart = tariffs.query('tariff == "smart"')['cost']
alpha = 0.05

results = st.ttest_ind(array_ultra, array_smart, equal_var = False)

print('p-значение:', results.pvalue)
print('среднее значение ultra:', round(array_ultra.mean(),2))
print('среднее значение smart:', round(array_smart.mean(),2))

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

### Средняя выручка пользователей из Москвы и выручка пользователей из других регионов

Гипотезы:

Ho - средняя выручка пользователей из Москвы и регионов равны,

H₁ - средняя выручка пользователей из Москвы и регионов различаются

```python
array_moscow = tariffs.query('region == "Москва"')['cost']
array_reg = tariffs.query('region == "не Москва"')['cost']
alpha = 0.05

results = st.ttest_ind(array_moscow, array_reg, equal_var = False)

print('p-значение:', results.pvalue)
print('среднее значение Москва:', round(array_moscow.mean(),2))
print('среднее значение Регионы:', round(array_reg.mean(),2))

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```
